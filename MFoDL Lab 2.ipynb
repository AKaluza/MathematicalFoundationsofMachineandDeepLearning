>
> *Two major traps to avoid today:*
> *1. **Correlation $\neq$ Causation:** Just because two variables move together (like ice cream sales and shark attacks) doesn't mean one causes the other. They might share a confounding variable (like summer heat).*
> *2. **Data Dredging (p-hacking):** If you torture the data long enough, it will confess to anything. Don't look for patterns just to fit your biases; look for generalizable truths.*

## 5. Summary (10 min)

### Key Takeaways
*   **Visuals drive decisions:** A simple histogram can tell you whether to apply a log transformation. A bar chart can reveal if a feature is predictive.
*   **Not all correlations are Pearson:** If your data is heavily skewed (like `Fare`), use Spearman correlation to find true relationships.
*   **EDA leads to Feature Engineering:** The goal of EDA is not just to draw pretty pictures, but to generate new columns (`family_size`, `is_child`) that help the model learn faster.

### The "Always Check in EDA" Checklist (12 Points)
Before you even import `sklearn`, make sure you have checked:
*   [ ] 1. Train/Test Split: Did I split the data to avoid leakage before deep EDA?
*   [ ] 2. Dataset Shape: How many rows and columns?
*   [ ] 3. Target Distribution: Is it balanced? Do I need SMOTE or class weights?
*   [ ] 4. Data Types: Are categorical variables stored as `object`/`category` and not `int`?
*   [ ] 5. Duplicates: Are there exact duplicated rows?
*   [ ] 6. Missing Values: How many NaNs per column? What's the imputation strategy?
*   [ ] 7. Missingness Patterns: Are NaNs random, or do they mean something (e.g., missing Cabin = no cabin)?
*   [ ] 8. Numeric Distributions: Are features normally distributed, or heavily skewed?
*   [ ] 9. Categorical Cardinality: Does a text column have 1000 unique values? (Needs grouping/encoding).
*   [ ] 10. Outliers: Are there impossible values (e.g., Age = 999) or extreme statistical outliers?
*   [ ] 11. Multicollinearity ($X$ vs $X$): Are two features 95% correlated? (Drop one to simplify the model).
*   [ ] 12. Predictive Power ($X$ vs $y$): Which features visually separate the target classes the best?

### 3 Control Questions (Ask the audience)
1.  **Question:** Why did we use `np.log1p()` instead of `np.log()` for the Fare column?
    *   *Answer:* Some passengers have a Fare of `0`. `log(0)` is mathematically undefined (returns `-inf`), but `log(1 + 0)` is safely `0`.
2.  **Question:** Why is it dangerous to perform target-based EDA (like finding the average survival rate per ticket type) on the entire dataset *before* splitting?
    *   *Answer:* It causes Data Leakage. The model will indirectly "learn" the distribution of the test set, giving you an overly optimistic evaluation score that will fail in production.
3.  **Question:** If a feature has a Pearson correlation of `0.05` with the target, but a Spearman correlation of `0.65`, what does this tell you?
    *   *Answer:* There is a strong relationship, but it is highly non-linear or heavily influenced by extreme outliers that Pearson cannot handle.
